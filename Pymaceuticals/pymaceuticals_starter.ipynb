```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import numpy as np
import seaborn as sns

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single DataFrame
combine_df = pd.merge(study_results, mouse_metadata, how = "left", on = "Mouse ID")

# Display the data table for preview
combine_df.head()
```

```python
# Checking the number of mice.
len(combine_df["Mouse ID"].unique())
```

```python
# Our data should be uniquely identified by Mouse ID and Timepoint
# Get the duplicate mice by ID number that shows up for Mouse ID and Timepoint.
duplicate_mice = combine_df.loc[combine_df.duplicated(subset = ["Mouse ID", "Timepoint"]), "Mouse ID"].unique()
duplicate_mice
```

```python
# Optional: Get all the data for the duplicate mouse ID.
duplicate_mice_df = combine_df.loc[combine_df["Mouse ID"] == "g989"]
duplicate_mice_df
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
clean_combine_df = combine_df[combine_df['Mouse ID'].isin(duplicate_mice) == False]
clean_combine_df
```

```python
# Checking the number of mice in the clean DataFrame.
len(clean_combine_df["Mouse ID"].unique())
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen
means = clean_combine_df.groupby("Drug Regimen")["Tumor Volume (mm3)"].mean()
medians = clean_combine_df.groupby("Drug Regimen")["Tumor Volume (mm3)"].median()
variances = clean_combine_df.groupby("Drug Regimen")["Tumor Volume (mm3)"].var()
stand_dev = clean_combine_df.groupby("Drug Regimen")["Tumor Volume (mm3)"].std()
sems = clean_combine_df.groupby("Drug Regimen")["Tumor Volume (mm3)"].sem()

# Use groupby and summary statistical methods to calculate the following properties of each drug regimen:
# mean, median, variance, standard deviation, and SEM of the tumor volume.
# Assemble the resulting series into a single summary DataFrame.
summary_table = pd.DataFrame({"Mean of Tumor Volume": means,
                             "Medians of Tumor Volume": medians,
                             "Variances of Tumor Volume": variances,
                             "Standard Deviation of Tumor Volume": stand_dev,
                             "SEM of Tumor Volume": sems
                             })
summary_table 
                            
                             
```

```python
# A more advanced method to generate a summary statistics table of mean, median, variance, standard deviation,
# and SEM of the tumor volume for each regimen (only one method is required in the solution)

# Using the aggregation method, produce the same summary statistics in a single line
summary_table = clean_combine_df.groupby("Drug Regimen").agg({"Tumor Volume (mm3)":["mean","median","var","std","sem"]})
summary_table
```

## Quartiles, Outliers and Boxplots

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:
# Capomulin, Ramicane, Infubinol, and Ceftamin

# Start by getting the last (greatest) timepoint for each mouse
max_time = clean_combine_df.groupby("Mouse ID").Timepoint.max().reset_index()

# Merge this group df with the original DataFrame to get the tumor volume at the last timepoint
merged_combine_df = pd.merge(max_time, clean_combine_df, on=["Mouse ID", "Timepoint"], how="left")

drugs = ["Capomulin", "Ramicane", "Infubinol", "Ceftamin"]
mask = merged_combine_df["Drug Regimen"].isin(drugs)
merged_combine_df = merged_combine_df.loc[mask].reset_index(drop=True)
merged_combine_df
```

```python
plt.figure(figsize=(12, 6))

# Create a basic plot
sns.boxplot(data= merged_combine_df, y="Tumor Volume (mm3)", x="Drug Regimen", color="magenta")

# Customizations

# Change colors
# Add in reference lines
# Add Labels/Title
plt.xlabel("Drug Regimen", fontfamily="Arial", fontsize=12)
plt.ylabel("Tumor Volume (mm3)", fontfamily="Arial", fontsize=12)
plt.title("Tumor Volume by Drug", fontweight="bold", fontsize=16, fontfamily="Arial")

# Add in a background grid
plt.grid(axis="y", linestyle="--", color="lightgrey", alpha=0.75)
# plt.xlim(0,20)

# Show/Save the Graph
plt.show()
```

```python
# Put treatments into a list for for loop (and later for plot labels)

treatments = merged_combine_df['Drug Regimen'].unique()

# Create empty list to fill with tumor vol data (for plotting)
tumor_volumes = []

# Calculate the IQR and quantitatively determine if there are any potential outliers.
# Iterate through each treatment
for treatment in treatments:
    # Locate the rows which contain mice on each drug and get the tumor volumes
    treatment_data = merged_combine_df.loc[merged_combine_df['Drug Regimen'] == treatment, 'Tumor Volume (mm3)']

    # add subset
    tumor_volumes.append(treatment_data)
    
    # Calculate IQR for outlier detection
    q1 = np.percentile(treatment_data, 25)
    q3 = np.percentile(treatment_data, 75)
    iqr = q3 - q1
    
    # Determine outliers using upper and lower bounds
    lower_bound = q1 - (1.5 * iqr)
    upper_bound = q3 + (1.5 * iqr)
    
    # Identify outliers
    outliers = treatment_data[(treatment_data < lower_bound) | (treatment_data > upper_bound)]
    print(f"Outliers for {treatment}: {outliers.tolist()}")
```

```python
tumor_volumes
```

```python
# Generate a box plot that shows the distribution of the tumor volume for each treatment group.
# Change Canvas Size
plt.figure(figsize=(12, 6))

plt.boxplot(tumor_volumes, tick_labels=treatments)

# Add Labels/Title
plt.xlabel("Drug Regimen", fontfamily="Arial", fontsize=12)
plt.ylabel("Tumor Volume (mm3)", fontfamily="Arial", fontsize=12)
plt.title("Tumor Volume by Drug", fontweight="bold", fontsize=16, fontfamily="Arial")

# Add in a background grid
plt.grid(axis="y", linestyle="--", color="pink", alpha=0.75)
# plt.xlim(0,20)

# Show/Save the Graph
plt.show()
```
